```python
description1 = []
feautres = []
price = []
originalprice = []
rating = []
```

```python
for page in range(1,27):
    url = 'https://www.flipkart.com/search?q=air+conditioners&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(page)
    req = requests.get(url)
    soup = BeautifulSoup(req.content)
    
    desc = soup.find_all('div',class_ = '_4rR01T')
    for i in range(len(desc)):
        description1.append(desc[i].text)
    
    commonclass = soup.find_all('ul',class_ = '_1xgFaf')
    for i in range(len(commonclass)):
        feautres.append(commonclass[i].text)
        
    mrp = soup.find_all('div' ,class_ ="_30jeq3 _1_WHN1")
    for i in range(len(mrp)):
        price.append(mrp[i].text)
    
    rat = soup.find_all('div',class_ = '_3LWZlK')
    for i in  range(len(rat)):
        rating.append(rat[i].text)
    
    orp = soup.find_all('div',class_='_3I9_wc _27UcVY')
    for i in range(len(orp)):
        originalprice.append(orp[i].text)
```
